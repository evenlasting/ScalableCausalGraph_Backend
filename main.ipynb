```python
from src.services.causal_graph_generator import CausalGraphGenerator
import matplotlib.pyplot as plt
import networkx as nx

from src.utils.draw_graph_topo import draw_dag

# load_dag = CausalGraphGenerator("study").generate_causal_graph()

import pickle

# with open('study_16_dag_file.pkl', 'wb') as f:
#     pickle.dump(load_dag, f)

# with open('dag_file.pkl','rb') as f:
#     load_dag = pickle.load(f)

with open('study_16_dag_file.pkl', 'rb') as f:
    load_dag = pickle.load(f)
```

```python
nx_graph = nx.DiGraph()

for node in load_dag.nodes():
    nx_graph.add_node(node)
for edge in load_dag.edges():
    nx_graph.add_edge(edge[0],edge[1])
draw_dag(nx_graph)
```

```python

from src.algorithm.louvain.community_louvain import best_partition,induced_graph
partition = best_partition(nx_graph)
graph = induced_graph(partition, nx_graph, allowCycle=False)
```
